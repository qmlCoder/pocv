```python
import numpy as np
import pyvista as pv
from pyvista.plotting.plotter import Plotter
import matplotlib.pyplot as plt
import matplotlib as mpl
import sys
sys.path.append("D:\code\pywfn")
from pywfn.base import Mol
from pywfn.reader import LogReader
from pywfn.atomprop import freeValence,directElectron
from pywfn.bondprop import piDM
from pywfn.maths import points_rotate
from draw3 import get_around,draw_mol,build_mesh
import matplotlib.pyplot as plt
import matplotlib as mpl
from pathlib import Path
from pywfn.utils import printer
printer.ifShell=False
```

```python
def get_dirs(angles):
    dirs:list[np.ndarray]=[]
    for a in angles:
        x=0
        y=np.sin(a)
        z=np.cos(a)
        dirs.append(np.array([x,y,z]))
    return dirs
```

```python

angs=np.linspace(0,np.pi*2,45)[:-1]
dirs=get_dirs(angs)
data=np.zeros(shape=(2,len(dirs),5)) #原子，方向，（负电荷、中电荷、正电荷、键级1，键级2）
atoms=[1,3]
bonds=[(2,3),(1,4)]
name='C4-OH'

molN=Mol(reader=LogReader(f"D:\BaiduSyncdisk\Articles\HFV\gfile\ok\SI\R-{name}-.log"))
mol0=Mol(reader=LogReader(f"D:\BaiduSyncdisk\Articles\HFV\gfile\ok\SI\R-{name}.log"))
molP=Mol(reader=LogReader(f"D:\BaiduSyncdisk\Articles\HFV\gfile\ok\SI\R-{name}+.log"))


calerN=directElectron.Calculator(molN)
caler0=directElectron.Calculator(mol0)
calerP=directElectron.Calculator(molP)

for a,atom in enumerate(atoms):
    for m,cal in enumerate([calerN,caler0,calerP]):
        cal.atoms=[atom]*len(dirs)
        cal.vects=[np.copy(dir) for dir in dirs]
        cal.multi=False
        cal.abs=False
        res=cal.calculate()
        data[a,:,m]=res

# caler=freeValence.Calculator(mol0)
caler=piDM.Calculator(mol0)
for a,atom in enumerate(atoms):
    # b1,b2=bonds[a]
    for b,bi in enumerate(bonds[a]):
        for d,dir in enumerate(dirs):
            caler.direct=dir
            caler.bond=[atom,bi]
            o1,o2=caler.calculate()
            data[a,d,3+b]=o1
np.save(f'数据/{name}.npy',data)
```

```python
def if_fill(i,j):
    ijs=[
        (1,2),(1,3),(1,4),
        # (1,7),(1,8),(1,9),
        ]
    for ti,tj in ijs:
        if i==ti and j==tj:
            return True
    return False
img=Image.new('RGB',(250,250),'white')
draw=ImageDraw.Draw(img)
b=50
r=4
for i in range(5):
    for j in range(5):
        x=i*b+r
        y=j*b+r
        if j in [2,3,4]:
            if i==1:
                color='#2b8a3e'
            else:
                color='#d3f9d8'
        else:
            color='#e9ecef'
        draw.rectangle((x,y,x+b-r*2,y+b-r*2),fill=color)
img.save('new.png')
```
