## 连烯pi电子分布

```python
import sys
sys.path.append('D:\code\pywfn')
import pywfn
print(pywfn)
import numpy as np
from pywfn.reader import LogReader
from pywfn.atomprop import piElectron,mullikenCharge
```

```python
piEcts=[]
muEcts=[]
for i in range(36):
    path=rf"D:\BaiduSyncdisk\gfile\scans\lianxi\lianxiScan\f{i+1:0>2}.log"
    reader=LogReader(path)
    mol=reader.mol
    caler=piElectron.Calculator(mol)
    elects=caler.calculate()
    piEcts.append([elects[i] for i in (0,2,4,7)])
    caler=mullikenCharge.Calculator(mol)
    elects=caler.calculate()
    muEcts.append([elects[i] for i in (0,2,4,7)])
```

```python
piEcts=np.array(piEcts)
muEcts=np.array(muEcts)
```

```python
np.save('数据/ectsPi-lianxi.npy',piEcts)
np.save('数据/ectsMu-lianxi.npy',muEcts)
```

## 取代联苯数据

```python
import sys
sys.path.append('D:\code\pywfn')
import numpy as np
from pywfn.readers import LogReader
from pywfn.atomprop import piElectron,mullikenCharge
from pywfn.bondorder import piDM,piSM,mayer
```

```python
piOrderP=np.zeros((36,4))
piOrderS=np.zeros((36,4))
OrderM=np.zeros((36,4))
bonds=[(10,11),(10,12),(10,3),(3,2)]
for i in range(36):
    path=rf"D:\BaiduSyncdisk\gfile\scans\lianbenR\lianbenRScan\f{i+1:0>2}.log"
    reader=LogReader(path)
    mol=reader.mol
    calerS=piSM.Calculator(mol)
    calerP=piDM.Calculator(mol)
    calerM=mayer.Calculator(mol)
    for j,(a1,a2) in enumerate(bonds):
        orderS=calerS.calculate(a1,a2)
        orderP=calerP.calculate(a1,a2)[0]
        orderM=calerM.calculate(a1,a2)
        piOrderS[i,j]=orderS
        piOrderP[i,j]=orderP
        OrderM[i,j]=orderM
```

```python
np.save('数据/取代联苯piOrderP.npy',piOrderP)
np.save('数据/取代联苯piOrderS.npy',piOrderS)
np.save('数据/取代联苯OrderM.npy',OrderM)
```
