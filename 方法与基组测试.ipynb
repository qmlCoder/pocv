```python
import sys;sys.path.append('d:/code/pywfn')
from pywfn.base import Mol
from pywfn.reader import LogReader
from pywfn.bondprop import order

root="D:/gfile/HFV/取代联苯扫描/wfns"
datas=np.zeros(shape=(6,6,36,4)) # 6种方法，6个基组，36个角度，4个键
bonds=[(10,11),(10,12),(3,10),(2,3)]
for m in range(6):
    for b in range(6):
        for a in range(36):
            path=rf"{root}/scan_m{m+1}_b{b+1}_lsp/{a+1}.log"
            mol=Mol(LogReader(path))
            caler=order.Calculator(mol)
            orders=caler.pi_pocv()
            vals=[]
            for a1,a2,val in orders:
                if (int(a1),int(a2)) in bonds:
                    vals.append(val)
            datas[m,b,a,:]=np.array(vals)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.axes._axes import Axes
import numpy.typing as tp

root=rf"D:\BaiduSyncdisk\Articles\HFV\gfile\取代联苯扫描\wfns"
# ms=['hf','b3lyp','b3lyp-d3','m062x','wb97xd','mp2']
ms=['HF','B3LYP','B3LYP-D3','M062X','$\omega $B97X-D','MP2']
bs=['STO-3G','3-21G','6-31G','6-31G(d)','6-31G(d,p)','6-311G(d,p)']
cs=['#2b9d8d','#025699','#f2a363','#e96d53']
xs=np.linspace(-44.96,-44.96+360,36)
xticks=[f'{x:.0f}°' if x<=180 else f'{x-360:.0f}°' for x in xs]
markers=['o','s','^','D']
plt.rc('font',family='Times New Roman',size=10)
for m in range(6):
    fig,axs=plt.subplots(2,3,figsize=(12,6))
    axs:list[Axes]=axs.flatten() # type: ignore
    plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.3)
    for b in range(6):
        orders=np.loadtxt(f'数据/方法基组测试/{m+1},{b+1}.npy')
        orders=orders[:,[3,2,0,1]].copy()
        axs[b].set_title(f'{ms[m]}/{bs[b]}')
        for o in range(4):
            order=orders[:,o]
            order=datas[m,b,:,o]
            axs[b].set_xticks(xs[::5],labels=xticks[::5])
            axs[b].plot(xs,order,marker=markers[o],color=cs[o],markersize=3,linewidth=1)
    fig.savefig(f'图像/方法基组测试/{m+1},{b+1}_.png',bbox_inches='tight')
```
