```python
root=rf"D:\BaiduSyncdisk\Articles\HFV\gfile\\wfns"
# ms=['hf','b3lyp','b3lyp-d3','m062x','wb97xd','mp2']
ms=['HF','B3LYP','B3LYP-D3','M062X','$\omega $B97X-D','MP2']
bs=['STO-3G','3-21G','6-31G','6-31G(d)','6-31G(d,p)','6-311G(d,p)']
cs=['#2b9d8d','#025699','#f2a363','#e96d53']
xs=np.linspace(-44.96,-44.96+360,36)
xticks=[f'{x:.0f}' if x<=180 else f'{x-360:.0f}' for x in xs]
plt.rc('font',family='Times New Roman',size=10)
for m in range(6):
    fig,axs=plt.subplots(2,3,figsize=(12,6))
    plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.3)
    for b in range(6):
        orders=np.loadtxt(f'//{m+1},{b+1}.npy')
        axs[b//3,b%3].set_title(f'{ms[m]}/{bs[b]}')
        for o in range(4):
            order=orders[:,o]
            axs[b//3,b%3].set_xticks(xs[::5],labels=xticks[::5])
            axs[b//3,b%3].plot(xs,order,marker='.',color=cs[o])
    fig.savefig(f'//{m+1},{b+1}.png',bbox_inches='tight')
```
